```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import metrics
from xgboost import XGBRegressor
```

```python
data = pd.read_csv("2022.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
data.isnull().sum()

```

```python
data["Happiness score"].mean()
data["Whisker-high"].mean()

```

```python
data["Happiness score"].fillna(data["Happiness score"].mean(),inplace=True)
data["Whisker-high"].fillna(data["Whisker-high"].mean(),inplace=True)
data["Whisker-low"].fillna(data["Whisker-low"].mean(),inplace=True)
data["Dystopia (1.83) + residual"].fillna(data["Dystopia (1.83) + residual"].mean(),inplace=True)
data["Explained by: GDP per capita"].fillna(data["Explained by: GDP per capita"].mean(),inplace=True)
data["Explained by: Social support"].fillna(data["Explained by: Social support"].mean(),inplace=True)
data["Explained by: Healthy life expectancy"].fillna(data["Explained by: Healthy life expectancy"].mean(),inplace=True)
data["Explained by: Freedom to make life choices"].fillna(data["Explained by: Freedom to make life choices"].mean(),inplace=True)
data["Explained by: Generosity"].fillna(data["Explained by: Generosity"].mean(),inplace=True)
data["Explained by: Perceptions of corruption"].fillna(data["Explained by: Perceptions of corruption"].mean(),inplace=True)


```

```python
x = data.drop(['Country','Happiness score','RANK'],axis=1)
y = data["Happiness score"]


```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
error_score = metrics.r2_score(y_train,train_data_prediction)
print("r score is -->",error_score)
```
